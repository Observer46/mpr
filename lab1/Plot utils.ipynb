```python
import matplotlib.pyplot as plt
import numpy as np
# from importlib import reload
# reload(plt)
# plt.xlabel("Hello World")

plt.rcParams['figure.figsize'] = [14, 8]
```

```python
def read_data_from_file(filename):
    data = np.loadtxt(filename, delimiter=' ')
#     print(data)
    msg_size, time, throughput = data[:,0], data[:,1], data[:,2]
    return msg_size, throughput
#     return k_data, t_data, add_data, mul_data, total_data

def plot_throughput(files, labels):
    for i, file in enumerate(files):
        data = read_data_from_file(file)
        plt.plot(data[0], data[1], label=labels[i])#, marker='o')
    
    plt.ylabel("Throughput [Mbit/s]")
    plt.xlabel("Message size [B]")
    plt.legend()
#     plt.ticklabel_format(style='plain')
    plt.savefig("throughput.png")#,style='plain')
    plt.show()
    
files = ["ssend_th_output_2.txt"]

plot_throughput(files, files)
# plot_add_mul(files, "float_ops")
```
